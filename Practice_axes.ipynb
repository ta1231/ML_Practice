```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import drive
data = pd.read_csv("/content/anomaly_86.csv")
```

```python
wavelength = np.array(data['wavelength'][:])
x = np.array(data['x'][:])
y = np.array(data['y'][:])
z = np.array(data['z'][:])
```

```python
plt.scatter(wavelength, x)
plt.title("x")
plt.show()
```

```python
plt.scatter(wavelength, y)
plt.title("y")
plt.show()
```

```python
plt.scatter(wavelength, z)
plt.title("z")
plt.show()
```

```python
plt.scatter(x, y)
plt.title("x-y")
plt.show()
```

```python
plt.scatter(x, z)
plt.title("x-z")
plt.show()
```

```python
plt.scatter(y, z)
plt.title("y-z")
plt.show()
```

```python
plt.xlabel("wavelength")
plt.ylabel("x")

plt.plot(wavelength, x)
plt.show()
```

```python
plt.xlabel("wavelength")
plt.ylabel("y")

plt.plot(wavelength, y)
plt.show()
```

```python
sum = x + y + z
```

```python
plt.scatter(wavelength, sum)
plt.title("x+y+z")
plt.show()
```

```python
plt.hist(sum)
plt.title("xyz_hist")
plt.show() 
```

```python
mean = np.average(sum)
std = np.std(sum)
mini = min(sum)
maxi = max(sum)
c = len(sum)
numOfBins = 10
```

```python
# import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25, 7),
                        facecolor='linen')
fig.suptitle("Anomaly",
 fontsize=30,
 fontfamily='monospace')

ax1 = fig.add_subplot(241)
ax2 = fig.add_subplot(242)
ax3 = fig.add_subplot(243)
ax4 = fig.add_subplot(244)
ax5 = fig.add_subplot(245)
ax6 = fig.add_subplot(246)
ax7 = fig.add_subplot(247)
ax8 = fig.add_subplot(248)

ax1.scatter(wavelength, x, color = 'blue')
ax1.set_title('x')
ax2.scatter(wavelength, y, color = 'blue')
ax2.set_title('y')
ax3.scatter(wavelength, z, color = 'blue')
ax3.set_title('z')
ax4.hist(sum, color = 'red')
ax4.set_title('xyz_hist')
ax4.text(s = 'Mean: ' + str(round(mean, 4)) + "\n"
+ "Std: " + str(round(std, 4)) + "\n"
+ "# of Bins: " + str(numOfBins) + "\n"
+ "Max: " + str(round(maxi, 4)) + "\n"
+ "Min: " + str(round(mini, 4)), x = -0.9, y = 10)

ax5.scatter(x, y, color = 'green')
ax5.set_title('x-y')
ax6.scatter(x, z, color = 'green')
ax6.set_title('x-z')
ax7.scatter(y, z, color = 'green')
ax7.set_title('y-z')
ax8.scatter(wavelength, sum, color = 'red', )
ax8.set_title('x+y+z')

plt.show()
```
